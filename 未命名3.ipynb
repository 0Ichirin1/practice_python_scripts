```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 折线图
# 数据准备
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
y = [5, 3, 4, 20, 18, 6, 7, 8, 9, 56, 35]
# 用Matplotlib画
plt.plot(x, y)
plt.show()
# 用Seaborn画
df = pd.DataFrame({"x":x, "y":y})
sns.lineplot(x="x", y="y", data=df)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 直方图
# 数据准备
a = np.random.randn(100)
s = pd.Series(a)
# 用Matplotlib画 
plt.hist(s)
plt.show()
# 用Seaborn画
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 条形图
# 数据准备
x = ["cat1", "cat2", "cat3", "cat4", "cat5"]
y = [5, 4, 8, 12, 7]
# 用Matplotlib画 
plt.bar(x, y)
plt.show()
# 用Seaborn画
sns.barplot(x, y)
plt.show()
```

```python
# 箱线图
# 数据准备 生成 10*4 维度数据
data = np.random.normal(size=(10, 4))
labels = ["A", "B", "C", "D"]
# 用Matplotlib画 
plt.boxplot(data, labels)
plt.show()
# 用Seaborn画
df = DataFrame(data, columns=labels)
sns.boxplot(data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
# 饼图
# 数据准备
nums = [25, 37, 33, 67, 6]
labels = ["High-school", "Bachelor", "Master", "Ph.d", "Others"]
# 用Matplotlib画 
plt.pie(x=nums, labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 热力图
# 数据准备
flights = sns.load_dataset("flights")
data = flights.pivot("year", "month", "passengers")
# 用Seaborn画
sns.heatmap(data)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties
# 蜘蛛图
# 数据准备
labels = np.array([u"推进","KDA",u"生存",u"团战",u"发育",u"输出"])
stats = [83, 62, 95, 67, 57, 88]
# 画图数据准备， 角度， 状态值
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats = np.concatenate((stats, [stats[0]]))
angles = np.concatenate((angles, [angles[0]]))
# 用Matplotlib画
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, "o-", linewidth=2)
ax.fill(angles, stats, alpha=0.25)
# 设置中文字体
# font = FontProperties(fname=r"", size=14)
# ax.set_thetagrids(angles*180/np.pi, labels, FontProperties)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 二元变量分布（散点图，核密度图，Hexbin图）
# 数据准备
tips = sns.load_dataset("tips")
print(tips.head(10))
# 用Seaborn画
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 探索二元成对关系
# 数据准备
iris = sns.load_dataset("iris")
# 用Seaborn画
sns.pairplot(iris)
plt.show()
```

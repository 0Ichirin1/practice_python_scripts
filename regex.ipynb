```python
import re

# 正则匹配文本内容
content = 'hello 123 4567 world_this is a regex demo'
print(len(content))
result = re.match('hello\s\d{3}\s\d{4}', content)
print(result)
print(result.group())
print(result.span())
```

```python
# 文本替换内容
content = 'hello 1234567 world_this is a regex demo'
result = re.match('hello\s(\d+)\sworld', content)
print(result)
print(result.group())
print(result.group(1))
print(result.span())
```

```python
# 文本替换内容
content = 'hello 1234567 world_this is a regex demo'
result = re.match('hello(.*)world', content)
print(result)
print(result.group())
print(result.group(1))
print(result.span())
```

```python
# 文本替换内容
content = 'hello 1234567 world_this is a regex demo'
result = re.match('^hello.*?(\d+).*demo$', content)
print(result)
print(result.group())
print(result.group(1))
print(result.span())
```

```python
# 文本替换内容
content = 'http://weibo.com/comment/kEraCN'
result1 = re.match('^http.*?comment/(.*?)', content)
result2 = re.match('^http.*?comment/(.*)', content)
print(result1)
print(result2)
print(result1.group())
print(result1.group(1))
print(result2.group())
print(result2.group(1))
```

```python
import re
content1 = '2020-12-15 12:00'
content2 = '2019-32-54 23:01'
content3 = '2098-31-97 63:32'
pattern = re.compile('\d{2}:\d{2}')
res1 = re.sub(pattern, '' , content1)
res2 = re.sub(pattern, 'two', content2)
res3 = re.sub(pattern, 'three', content3)
print(res1, res2, res3)
```

```python
html = '''
<div class="wrap">
    <div id="container">
        <ul class="list">
             <li class="item-0">first item</li>
             <li class="item-1"><a href=" ">second item</a ></li>
             <li class="item-0 active"><a href="link3.html"><span class="bold">third item</span></a ></li>
             <li class="item-1 active"><a href="link4.html">fourth item</a ></li>
             <li class="item-0"><a href="link5.html">fifth item</a ></li>
         </ul>
     </div>
 </div>
'''
doc = pq(html)
li = doc('.list .item-0.active a')
print(li, type(li))
print(li.attr('href'))
print(str(li.siblings()).strip())
```

```python
html = '''
<div class="wrap">
    <div id="container">
        <ul class="list">
             <li class="item-0">first item</li>
             <li class="item-1"><a href=" ">second item</a ></li>
             <li class="item-0 active"><a href="link3.html"><span class="bold">third item</span></a ></li>
             <li class="item-1 active"><a href="link4.html">fourth item</a ></li>
             <li class="item-0"><a href="link5.html">fifth item</a ></li>
         </ul>
     </div>
 </div>
'''
doc = pq(html)
lis = doc('li').items()
print(type(lis))
for li in lis:
    print(li, type(li))
```

```python
html = '''
<div class="wrap">
    <div id="container">
        <ul class="list">
             <li class="item-0">first item</li>
             <li class="item-1"><a href=" ">second item</a ></li>
             <li class="item-0 active"><a href="link3.html"><span class="bold">third item</span></a ></li>
             <li class="item-1 active"><a href="link4.html">fourth item</a ></li>
             <li class="item-0"><a href="link5.html">fifth item</a ></li>
         </ul>
     </div>
 </div>
'''
doc = pq(html)
li = doc('li:first-child')
print(li)
li = doc('li:last-child')
print(li)
li = doc('li:nth-child(2)')
print(li)
li = doc('li:gt(2)')
print(li)
li = doc('li:nth-child(2n)')
print(li)
li = doc('li:contains(second)')
print(li)
```

```python
import pymongo
client = pymongo.MongoClient(host='localhost', port=27017)
db = client.test   # 建库
collection = db.students # 建表，把表赋给一个collection对象
student = {
    'id':'20180101',
    'name':'jordan',
    'age':28,
    'gender':'male'
}
res = collection.insert_one(student)
print(res)
print(res.inserted_id)
```
